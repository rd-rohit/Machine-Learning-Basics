```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ex2data1.txt", header=None)
df.head()
```

```python
X = df.iloc[:, :-1]
X.head()
```

```python
y = df.iloc[:,-1]
y.head()
```

```python
X = np.c_[np.ones((X.shape[0], 1)), X]
X
```

```python
y = y[:,np.newaxis]
y
```

```python
theta = np.zeros((X.shape[1], 1))
theta
```

```python
def hypothesis(theta, X):
    z = np.dot(X, theta)
    return 1/(1+np.exp(-(z))) - 0.0000001
```

```python
h = hypothesis(theta, X)
np.dot(X.T, (h-y))
```

```python
def costFunc(theta, X , y):
    y1 = hypothesis(theta, X)
    m = X.shape[0]
    return -(1/m) * np.sum(y*np.log(y1) + (1-y)*np.log(1-y1))
```

```python
def gradient(theta, X, y):
    m = X.shape[0]
    h = hypothesis(theta, X)
    return 1/m * np.dot(X.T, (h-y))
```

```python
from scipy.optimize import minimize, fmin_tnc
def fit(X,y,theta):
    para = fmin_tnc(func=costFunc, x0=theta, fprime=gradient, args=(X, y.flatten()))
    return para[0]
parameters = fit(X, y , theta)
```

```python
parameters
```

```python
h = hypothesis(parameters, X)
h
```

```python
def predict(h):
    h1 = []
    for i in h:
        if i>=0.5:
            h1.append(1)
        else:
            h1.append(0)
    return h1
y_pred = predict(h)
accuracy = 0
for i in range(0, len(y_pred)):
    if y_pred[i] == y[i]:
        accuracy += 1
accuracy/len(y)
```
